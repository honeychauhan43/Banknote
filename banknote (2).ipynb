import all the library which we will use

```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
import seaborn as sns
```

datasets link

```python
data=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt")
```

```python
data
```

```python
sns.pairplot(data) # data visulization
```

```python
data.describe() 
```

```python
#split the datasets


x = data.iloc[:,0:4].values
print(x)



y = data.iloc[:,-1].values
print(y)
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
y = pd.DataFrame(data = y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.22,random_state=50)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
pred = lg.predict(x_test)
print(pred)
```

```python
y_test
```

```python
print('acc score:',accuracy_score(pred,y_test))
```

```python
print(confusion_matrix(pred,y_test))
print(classification_report(pred,y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
pred = gnb.predict(x_test)
print(accuracy_score(pred,y_test))
```

```python
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfr= RandomForestClassifier(n_estimators = 100, random_state=48)
rfr = RandomForestClassifier()
rfr.fit(x_train,y_train)
pred = rfr.predict(x_test)
print(accuracy_score(pred,y_test))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn import model_selection

scores= model_selection.cross_val_score(rfr,x_train,y_train,scoring = 'accuracy',cv = 50)
banknote = scores,scores.mean()
```
